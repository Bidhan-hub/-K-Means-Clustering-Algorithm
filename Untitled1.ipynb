```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('/content/sample_data/income.csv')  # Loading csv file
print(df.head())
```

```python
#  Initial Scatter Plot (Age vs Income)
plt.scatter(df['Age'], df['Income($)'])
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Scatter plot of Age vs Income($)')
plt.show()
```

```python
# Ignore 'Name' column because it has string
X = df[['Age', 'Income($)']].values
```

```python
#Fit KMeans with cluster = 3
kmeans = KMeans(n_clusters=3, random_state=42)
y_kmeans = kmeans.fit_predict(X)
```

```python
# Plot clusters
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis')
plt.xlabel("Age")
plt.ylabel("Income")
plt.title("KMeans Clustering (Unscaled)")
plt.show()
```

```python
# Feature scaling using MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Fit KMeans again with cluster = 3 on scaled data
kmeans_scaled = KMeans(n_clusters=3, random_state=42)
y_kmeans_scaled = kmeans_scaled.fit_predict(X_scaled)
```

```python
# Plot clusters with scaled data and centroids
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y_kmeans_scaled, cmap='viridis', alpha=0.6)
centroids = kmeans_scaled.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.xlabel("Age (scaled)")
plt.ylabel("Income (scaled)")
plt.title("KMeans Clustering (Scaled) with Centroids")
plt.legend()
plt.show()
```

```python
# Use ELBO method (SSE) to find optimal K
sse = []
k_range = range(1, 10)
for k in k_range:
    kmeans_elbo = KMeans(n_clusters=k, random_state=42)
    kmeans_elbo.fit(X_scaled)
    sse.append(kmeans_elbo.inertia_)  # inertia_ is SSE
```

```python
# Plot SSE vs K
plt.plot(k_range, sse, marker='o')
plt.xlabel("Number of clusters K")
plt.ylabel("SSE (Sum of Squared Errors)")
plt.title("Elbow Method for Optimal K")
plt.show()
```
